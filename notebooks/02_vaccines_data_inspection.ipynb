## Nulls and duplicates

### Null values

```python
vaccines.isna().sum()/len(vaccines) * 100
```

```python
vaccines = vaccines.sort_values(['Country', 'Date'])
```

```python
cols_to_ffill = [i for i in vaccines.columns if i not in ['Country', 'Date']]
vaccines[cols_to_ffill] = (vaccines.groupby('Country')[cols_to_ffill].ffill())
```

```python
vaccines.isna().sum()/len(vaccines) * 100
```

After applying forward fill to the selected columns, the proportion of missing values remained unchanged. This suggests that the missing data is not due to reporting gaps, but rather because these values did not exist at earlier stages (for example, before certain vaccination measures were introduced).

Therefore, these missing values are considered to be structurally missing. To preserve the integrity of the dataset and avoid introducing bias, they have been retained as they are.

This decision ensures that the dataset accurately reflects real-world conditions without artificially altering the underlying data.

### - Column Names

In this dataset, the `Country` column corresponds to the `Country_code` column in the ***COVID cases and deaths*** dataset.
The same happens with `Date` which is named as `Date_reported` in the other dataset. To ensure consistency and facilitate the merge process, the `Country` and `Date` columns will be renamed to `Country_code` and `Date_reported`.


```python
vaccines.rename(columns={
    'Country': 'Country_code',
    'Date': 'Date_reported'}, inplace=True)
```
